- label:
    - 0 - nothing
    - 1 - stand
    - 2 - sit
    - 3 - lie
    - 4 - go
    - 5 - run

```python
train_y["label"]=0
for i in range(train_y.shape[0]):
    if train_y["nothing"][i]==1:
        train_y["label"][i]= 0
    elif train_y["stand"][i]==1:
        train_y["label"][i]= 1
    elif train_y["sit"][i]==1:
        train_y["label"][i]= 2
    elif train_y["lie"][i]==1:
        train_y["label"][i]= 3
    elif train_y["go"][i]==1:
        train_y["label"][i]= 4
    elif train_y["run"][i]==1:
        train_y["label"][i]= 5
```

```python
train_y.head()
```

```python
y_train=train_y.drop(["nothing","stand","sit","lie","go","run","id"],axis=1)
# we can also use
#train_y.drop(["nothing","stand","sit","lie","go","run"],axis=1,inplace=True)
```

```python
X_train=train_x.drop(["id"],axis=1)
X_test=test_x.drop(["id"],axis=1)
```

```python
y_train=y_train["label"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)
knn_predictions = knn.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train)
dtree_predictions = dtree_model.predict(X_test)
```

```python
%%time
from sklearn.svm import SVC
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train)
svm_predictions = svm_model_linear.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(class_weight="balanced",random_state=42,max_depth=5,max_features=1,n_estimators=10)
RFC.fit(X_train, y_train)
rfc_predictions=RFC.predict(X_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(n_estimators=50,learning_rate=1)
model = abc.fit(X_train, y_train)
abc_predictions = model.predict(X_test)
```

```python
algos=["knn","dtree","svm","gnb","rfc","abc"]
preds=[knn_predictions,dtree_predictions,svm_predictions,gnb_predictions,rfc_predictions,abc_predictions]
a=0
for i in preds:
    submission["label"]= i
    submission.to_csv("{}.csv".format(algos[a]),index=False)
    a+=1
```
